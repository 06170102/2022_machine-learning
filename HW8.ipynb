```python
#10 flod avg scores , k最佳值 weight何者較好 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('phone.csv')
df.head(5)
```

```python
df1=df[['battery_power','clock_speed','int_memory','m_dep','mobile_wt','n_cores','fc','pc','px_height','px_width','ram','sc_h','sc_w','talk_time']] #先將數值型資料做標準化處理
```

```python
from sklearn.preprocessing import StandardScaler
df1=StandardScaler().fit(df1).transform(df1) #將數值型資料進行標準化

df1
df1=pd.DataFrame(df1)
```

```python
df2=df[['blue','dual_sim','four_g','three_g','touch_screen','wifi']] #類別型資料
```

```python
df2=pd.get_dummies(df2,columns=['blue','dual_sim','four_g','three_g','touch_screen','wifi'],drop_first=True) #創立虛擬變數
```

```python
res = pd.concat([df1,df2],axis=1) #合併data
```

```python
X=np.array(res)
y=df[['price_range']] 
y=np.array(y)
y=np.ravel(y) #flattern y
print(y)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
scores=cross_val_score(knn,X,y,cv=10,scoring='accuracy')
print(scores)
```

```python
print(scores.mean()) #在k=5下 knn 10fold的平均accuracy
```

```python
from sklearn.model_selection import GridSearchCV #一定要cross validation
```

```python
k_range=list(range(1,100)) #krange1-100
weight_options=['uniform','distance']
```

```python
param_grid=dict(n_neighbors=k_range,weights=weight_options)
print(param_grid)
```

```python
grid=GridSearchCV(knn,param_grid,cv=10,scoring='accuracy')
grid.fit(X,y)
```

```python
pd.DataFrame(grid.cv_results_)[['params','mean_test_score','std_test_score']]
```

```python
print(grid.best_score_)
print(grid.best_params_)
```
