```python
#HW2 不同化學成分預測玻璃 描述dataset 用knn logistic 測testing accuracy k=3,5 比較三個model的accuracy
```

```python
import pandas as pd
import numpy as np  
```

```python
glass=pd.read_csv('glass.csv') #匯入glass csv
```

```python
print(glass) #資料集 有214筆觀測值 9個化學元素含量(features) 玻璃種類(target)可分為七個不同種類。目標：用化學元素含量來預測玻璃的種類
```

```python
target=glass[['Type']] #將type 設為target
element=glass[['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']] #將9個不同化學元素設為features，'RI','Na','Mg','Al','Si','K','Ca','Ba','Fe'
```

```python
element=element.values #將pd轉為array
```

```python
print(element)
```

```python
target=target.values #將pd轉為array
```

```python
print(target)
```

```python
target=np.concatenate(target) #更改array的形狀
print(target)
```

```python
print(type(element)) #檢查資料型態是否改對 ndarray正確
print(type(target))
```

```python
print(element.shape) #檢查shape 214筆 9feature 正確
```

```python
print(target.shape) #檢查shape
```

```python
X=element #將features設為Ｘ
```

```python
y=target #將target設為y
```

```python
print(X.shape)
print(y.shape) #確認shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0) #分割訓練集 測試集,訓練集70% 測試30%，random state確保結果可複製
```

```python
print(X_train.shape) #確認切分後的資料 訓練集149筆 測試集 65筆
print(X_test.shape)
```

```python
#knn
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier

```

```python
neigh = KNeighborsClassifier(n_neighbors=3) #設定參數k為3
```

```python
neigh.fit(X_train,y_train) #開始train data
```

```python
y_pred=neigh.predict(X_test) #測試集丟入training model,得出預測y值
print(y_pred)
```

```python
print(metrics.accuracy_score(y_test,y_pred)) #k=3的accuracy=0.646
```

```python
#knn 5
```

```python
knn = KNeighborsClassifier(n_neighbors=5) #設定參數k為5
```

```python
knn.fit(X_train,y_train) #開始train data
```

```python
y_pred=knn.predict(X_test) #測試集丟入training model,得出預測y值
print(y_pred)
```

```python
print(metrics.accuracy_score(y_test,y_pred)) #k=5的accuracy=0.661 較高
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression

log=LogisticRegression()

```

```python
log.fit(X_train,y_train) #開始train data
```

```python
y_pred=log.predict(X_test) #測試集丟入training model,得出預測y值
print(y_pred)
```

```python
print(metrics.accuracy_score(y_test,y_pred)) #logistic準確率僅0.553
```

```python
#accuracy比較 k=5>k=3>logistic
```
